# Uncertainty of the sum of two variables

```python
import numpy as np
import matplotlib.pyplot as plt
from tabulate import tabulate
%matplotlib inline
```

```python
np.random.seed(42)

n = 500000
x = np.random.uniform(-0.5,0.5,n)
y = np.random.uniform(-0.5,0.5,n)
z = x+y

width = 0.02
my_bins = np.arange(-1.5,1.5,width)

fig = plt.figure(figsize=(5,5))
axx = fig.add_subplot(3,1,1)
axy = fig.add_subplot(3,1,2)
axz = fig.add_subplot(3,1,3)

axx.hist(x,bins=my_bins,alpha=0.5,edgecolor='black')
axx.set_xticks([])
axx.set_ylabel('x counts')


axy.hist(y,bins=my_bins,alpha=0.5,edgecolor='black')
axy.set_xticks([])
axy.set_ylabel('y counts')


axz.hist(z,bins=my_bins,alpha=0.5,edgecolor='black')
axz.set_xlabel('value')
axz.set_ylabel('z counts')

plt.subplots_adjust(hspace=0.0)

plt.show()

#make a table to print
index = ['x', 'y', 'z']
columns = ['mean', 'stdev', 'var']
data = [[np.mean(x),np.mean(y),np.mean(z)],
        [np.std(x),np.std(y),np.std(z)],
        [np.var(x),np.var(y),np.var(z)]]
data = [[row[i] for row in data] for i in range(len(data[0]))]
table = tabulate(data, headers=columns, showindex=index, tablefmt='fancy_grid')
print(table)
```
